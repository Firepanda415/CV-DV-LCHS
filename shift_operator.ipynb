```python
import numpy as np
import scipy
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator
```

```python
def t_mat(M: int):
    mat = 2*np.identity(M, dtype=int)
    for j in range(0,M-2+1):
        mat[j+1,j] = -1
        mat[j,j+1] = -1
    return mat

def u_inc(M: int):
    mat = np.zeros((M, M), dtype=int)
    for j in range(0,M-2+1):
        mat[j+1,j] = 1
    mat[0,M-1] = 1
    return mat

def proj_d(M: int):
    mat = np.zeros((M, M), dtype=int)
    for j in range(0,M-2+1):
        mat[j,j] = 1
    return mat

def s_hat(M: int):
    mat = np.zeros((M, M), dtype=int)
    for j in range(0,M-2+1):
        mat[j+1,j] = 1
    return mat

    
```

```python
M = 4

u_inc(M)
```

Is $U_{inc}$ unitary?

```python
norm_diff1 = np.linalg.norm( u_inc(M).dot(u_inc(M).T) - u_inc(M).T.dot(u_inc(M))) 
norm_diff2 = np.linalg.norm( u_inc(M).dot(u_inc(M).T) - np.identity(M)) 

np.isclose(norm_diff1,0) and np.isclose(norm_diff2,0)
```
